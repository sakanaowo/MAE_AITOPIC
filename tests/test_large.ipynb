```python
import torch
from mae.mae_large import *

def test():
    model = mae_large_vit_large_patch16()
    
    params = sum(p.numel() for p in model.parameters())
    print(f'Total parameters: {params/1e6:.2f}M')
    
    x = torch.randn(2, 3, 224, 224)
    loss, pred, mask = model(x, mask_ratio=0.75)
    
    print(f"Input shape: {x.shape}")
    print(f"Loss: {loss.item():.4f}")
    print(f"Prediction shape: {pred.shape}")
    print(f"Mask ratio: {mask.mean().item():.4f}")
    print(f"Mask shape: {mask.shape}")
    
    print(f"Estimated GPU memory usage during training: {torch.cuda.memory_allocated()/1e9:.2f} GB")
    
if __name__ == "__main__":
    test()
    
```
